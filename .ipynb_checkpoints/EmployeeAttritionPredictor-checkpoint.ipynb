```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
data.shape
```

```python
# data.info()
data.head()
# data.describe()
```

```python
data['Attrition'] = data['Attrition'].map({'Yes':1, 'No':0})
```

```python
cols = data.select_dtypes(include='str').columns
cols
```

```python
le = LabelEncoder()
for col in cols:
    data[col] = le.fit_transform(data[col])
data.head()
```

```python
data.columns
```

```python
data.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1,  inplace=True)
```

```python
data.head()
```

```python
print(data['Attrition'].value_counts())
```

```python
sns.countplot(x='Department', hue='Attrition', data=data)
```

```python
corr = data.corr()
sns.heatmap(corr, cmap='coolwarm')
```

```python
x=data.drop(['Attrition'], axis=1)
y=data['Attrition']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=35)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
model=RandomForestClassifier(n_estimators=100, random_state=55)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
print('Classificatin report: ')
print(classification_report(y_test, y_pred))
```

```python
print('Accuracy: ')
print(accuracy_score(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import pandas as pd

importances = pd.Series(model.feature_importances_)
top_features = importances.sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
top_features.plot(kind='barh', color='steelblue')
plt.title('Top 10 Features Driving Employee Attrition')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.savefig('feature_importance.png')
plt.show()
```
